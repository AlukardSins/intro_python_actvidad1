* El constructor de la clase inicializa las coordenadas de esta (x,y) en el valor asignado por el usuario si se le pasan argumentos o en 0 si no se le pasan argumentos
* Para esta clase se definieron los siguientes métodos:
  * **distance**: Calcula la distancia de un punto a otro.
  * **set_location**: Asigna una posición determinada al punto.
  * **distance_from_origin**: Calcula la distancia del punto al origen de coordenadas.
  * **print**: Imprime las coordenadas siguiendo el siguiente formato: (x, y)
  
Dado el diagrama UML anterior hacer lo siguiente:
1. Codificar la clase en python.
2. Definir tres objetos tipo ```Point```: ```A(0,0)```,  ```B(-3,4)```, ```C(5,8)```.
3. Imprimir los tres puntos anteriormente creados.
4. Cambiar las coordenadas de A de tal modo que la componentes x de A sea la componente x de C y la componente y de A sea la componente y de B.
5. Imprimir nuevamente los puntos.
6. Calcular las distancias de AB, BC, AC y almacenarlas en las variables d_AB, d_BC y d_AC respectivamente. Luego, imprimir estos valores.
7. Calcule la distancia de A hasta el origen e imprimala.


````python
"""
Punto 1: Definicion de la clase punto
"""

from math import sqrt

# Clase Point
class Point:  
    
    # Constructor  
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y

    # Metodos  
    def set_location(self, x, y):
        self.x = x    
        self.y = y  
        
    def distance_from_origin(self):
        return sqrt(self.x * self.x + self.y * self.y)
    
    def distance(self, other):    
        dx = self.x - other.x    
        dy = self.y - other.y    
        return sqrt(dx**2 + dy**2)
    
    def print(self):
        print("(" + str(self.x) + ", " + str(self.y) + ")")
        

"""
Punto 2: A(0,0), B(-3,4), C(5,8)
"""
A = Point()
B = Point(-3,4)
C = Point(5,8)

"""
Punto 3: Impresion de los puntos A, B, C
"""
print("Coordenadas de los puntos A, B y C")
print("A:")
A.print()
print("B:")
B.print()
print("C:")
C.print()

"""
Punto 4: A.x = C.x y A.y = B.y
"""
A.set_location(C.x,B.y)

"""
Punto 5: Impresion de los puntos A, B, C
"""
print()
print("Nuevas coordenadas de los puntos A, B y C")
print("A:")
A.print()
print("B:")
B.print()
print("C:")
C.print()

"""
Punto 6: Calculo e impresión de las distancias
"""
d_AB = A.distance(B)
d_BC = B.distance(C)
d_AC = A.distance(C)
print()
print("Distancia entre A y B: " + str(d_AB))
print("Distancia entre B y C:",d_BC)
print("Distancia entre C y A: %s" %(d_AC))

"""
Punto 7: Calculo de la distancia de A hasta el origen
"""
print()
print("Distancia entre A y el origen es: " + str(A.distance_from_origin()))
````

````python
# Definicion de la clase Temperatura

class Temperature:

  def __init__(self, temperature = 75, subClass = 'Fahrenheit'):
      self.temperature = temperature
      self.subClass = subClass

  def str(self):
    return print('{}° {}'.format(self.temperature, self.subClass))
  
  def convertToFahren(self):
    pass

  def convertToCelcius(self):
    pass
  
  def convertToKelvin(self):
    pass

  def aboveFreezing(self):
    pass

# Definicion de subclases

# Fahrenheit
class Fahrenheit(Temperature):

  def __init__(self, temperature):
    super().__init__(temperature, 'Fahrenheit')

  def str(self):
    return super().str()

  def aboveFreezing(self):
    return self.temperature > 32

  def convertToFahren(self):
    return Fahrenheit(self.temperature)

  def convertToCelcius(self):
    return Celcius(5/9 * (self.temperature - 32))

  def convertToKelvin(self):
    return Kelvin(5/9 * (self.temperature - 32) + 273.15)

# Celcius
class Celcius(Temperature):

  def __init__(self, temperature):
    super().__init__(temperature , 'Celcius')
  
  def str(self):
    return super().str()
  
  def aboveFreezing(self):
    return self.temperature > 0

  def convertToFahren(self):
    return Fahrenheit((9/5 * self.temperature) + 32)
  
  def convertToCelcius(self):
    return Celcius(self.temperature)
  
  def convertToKelvin(self):
    return Kelvin(self.temperature + 273.15)

#Kelvin
class Kelvin(Temperature):

  def __init__(self, temperature):
    super().__init__(temperature, 'Kelvin')
  
  def str(self):
    return super().str()

  def aboveFreezing(self):
    return self.temperature > 273.15
  
  def convertToFahren(self):
    return Fahrenheit(9/5 * (self.temperature - 273.15) + 32)

  def convertToCelcius(self):
    return Celcius(self.temperature - 273.15)
  
  def convertToKelvin(self):
    return Kelvin(self.temperature)

# Definir arreglos a usar

# Temperaturas en 0
F0 = Fahrenheit(32)
C0 = Celcius(0)
K0 = Kelvin(273.15)

# Temperaturas 1° sobre 0
FA = Fahrenheit(33)
CA = Celcius(1)
KA = Kelvin(274.15)

# Temperaturas ambiente ideales
FI = Fahrenheit(80)
CI = Celcius(27)
KI = Kelvin(300)

# Listas
ListaCeros = [F0, C0, K0]
ListaSobreCero = [FA, CA, KA]
ListaIdeales = [FI, CI, KI]

ListaUnida = ListaCeros + ListaSobreCero + ListaIdeales

# Imprimir si esta por encima del punto de congelacion

print('Individuales')
for i in ListaUnida:
  if i.aboveFreezing():
    print('{}° {} is above freezin\n'.format(i.temperature, i.subClass))
  else:
    print('{}° {}\n'.format(i.temperature, i.subClass))

# Convertir y luego imprimir si esta por encima del punto de congelacion

## Fahrenheit
print('\nFahrenheit')
for i in ListaUnida:
  i = i.convertToFahren()
  if i.aboveFreezing():
    print('{}° is above freezin'.format(i.temperature))
  else:
    print('{}°'.format(i.temperature))

## Celcius
print('\nCelcius')
for i in ListaUnida:
  i = i.convertToCelcius()
  if i.aboveFreezing():
    print('{}° is above freezin'.format(i.temperature))
  else:
    print('{}°'.format(i.temperature))

## Kelvin
print('\nKelvin')
for i in ListaUnida:
  i = i.convertToKelvin()
  if i.aboveFreezing():
    print('{}° is above freezin'.format(i.temperature))
  else:
    print('{}°'.format(i.temperature))

````

````python
# Definicion de la clase Fraccion

from fractions import Fraction

class Fraccion:

  def __init__(self, fraccion):
    try:
      splitted = fraccion.split('/')
      if(splitted[1] == '0'):
        print('Denominador = 0, Division por cero se definira como 0')
        self.fraccion = 0
      else:
        self.fraccion = Fraction(fraccion)
    except:
      self.fraccion = Fraction(fraccion + "/1")

  def __str__(self):
    return str(self.fraccion)

  def __add__(self, otro):
    resultado = self.fraccion + otro.fraccion
    return Fraccion(str(resultado))

  def __sub__(self, otro):
    resultado = self.fraccion - otro.fraccion
    return Fraccion(str(resultado))

  def __mul__(self, otro):
    resultado = self.fraccion * otro.fraccion
    return Fraccion(str(resultado))
  
  def __truediv__(self, otro):
    resultado = self.fraccion / otro.fraccion
    return(str(resultado))

#Pruebas
Primera = Fraccion(input('Ingrese la primera fraccion, recordar que el simbolo de division es "/".\n'))
Segunda = Fraccion(input('Ingrese la segunda fraccion, recordar que el simbolo de division es "/".\n'))

print('\n')

print("Primera = {}".format(Primera))
print("Segunda = {}".format(Segunda))

print('\n')

suma = Primera + Segunda
print('La suma es = {}\n'.format(suma))

resta = Primera - Segunda
print('La resta (Primera - Segunda) es = {}\n'.format(resta))
resta = Segunda - Primera
print('La resta (Segunda - Primera) es = {}\n'.format(resta))

multiplicacion = Primera * Segunda
print('La multiplicacion es = {}\n'.format(multiplicacion))

if Primera != 0 or Segunda != 0:
  # Primera / Segunda
  if Primera.fraccion == 0:
    print('Dividendo = 0, Division de cero se definira como 0')
    print('La division (Primera / Segunda) es = 0\n')
  elif Segunda.fraccion == 0:
    print('Denominador = 0, Division por cero se definira como 0')
    print('La division (Primera / Segunda) es = 0\n')
  else:
    division = Primera / Segunda
    print('La division (Primera / Segunda) es = {}\n'.format(division))

  # Segunda / Primera
  if Segunda.fraccion == 0:
    print('Dividendo = 0, Division de cero se definira como 0')
    print('La division (Segunda / Primera) es = 0\n')
  elif Primera.fraccion == 0:
    print('Denominador = 0, Division por cero se definira como 0')
    print('La division (Segunda / Primera) es = 0\n')
  else:
    division = Segunda / Primera
    print('La division (Segunda / Primera) es = {}\n'.format(division))
else:
  print('Ambos factores son 0')
````
